```python
def processar_populacao_notebook(caminho_csv, nome_faixa):
    """Lê o dataset, limpa dados, consolida colunas do mesmo ano e faz unpivot."""
    df = pd.read_csv(caminho_csv, encoding='utf-8')
    
    # Separar a coluna de texto (Territórios) das numéricas (Anos)
    df_territorios = df[["Territórios"]].copy()
    df_numerico = df.drop(columns=["Territórios"])
    
    # Converter para número primeiro (enquanto os nomes são únicos)
    for col in df_numerico.columns:
        df_numerico[col] = pd.to_numeric(df_numerico[col].astype(str).str.replace(' ', ''), errors='coerce').fillna(0)
        
    # Limpar os sufixos das colunas usando Regex
    novas_colunas = {}
    for col in df_numerico.columns:
        ano_limpo = re.search(r'\d{4}', col)
        if ano_limpo: 
            novas_colunas[col] = ano_limpo.group(0)
    
    df_numerico = df_numerico.rename(columns=novas_colunas)
    
    # SOMAR colunas duplicadas do mesmo ano
    df_numerico = df_numerico.T.groupby(level=0).sum().T
    
    # Juntar novamente os territórios e agregar
    df_final = pd.concat([df_territorios, df_numerico], axis=1)
    df_agrupado = agregar_por_distrito(df_final)
    
    # Transformar para formato longo (Unpivot / Melt)
    colunas_anos = [c for c in df_agrupado.columns if c != "Territórios"]
    df_longo = df_agrupado.melt(
        id_vars=["Territórios"], 
        value_vars=colunas_anos, 
        var_name="Ano", 
        value_name="Total_Populacao"
    ).rename(columns={"Territórios": "Distrito"})
    
    df_longo["Faixa_Etaria"] = nome_faixa
    return df_longo

# --- EXECUTAR POPULAÇÃO ---
ficheiros_populacao = {
    "Dataset 0-14.csv": "00-14 anos", "Dataset 15-24.csv": "15-24 anos",
    "Dataset 25-49.csv": "25-49 anos", "Dataset 50-59.csv": "50-59 anos",
    "Dataset 60-74.csv": "60-74 anos", "Dataset 75-84.csv": "75-84 anos",
    "Dataset 85+.csv": "85+ anos"
}

os.makedirs("data/processed", exist_ok=True)
lista_dfs = []

for ficheiro, nome_faixa in ficheiros_populacao.items():
    caminho = f"data/raw/{ficheiro}" 
    if os.path.exists(caminho):
        try:
            df_tratado = processar_populacao_notebook(caminho, nome_faixa)
            lista_dfs.append(df_tratado)
            print(f"✔ População: {ficheiro} processado.")
        except Exception as e:
            print(f"❌ Erro ao processar {ficheiro}: {e}")
    else:
        print(f"❌ Não encontrado: {caminho}")

if lista_dfs:
    fact_populacao = pd.concat(lista_dfs, ignore_index=True)
    fact_populacao.to_csv("data/processed/Fact_Populacao.csv", index=False)
    print("\n🎉 Ficheiro 'Fact_Populacao.csv' consolidado e guardado!")
```

```python
def encontrar_linha_cabecalho(caminho_csv, termo_busca="Territ"): 
    """Encontra a linha onde começam os dados reais, ignorando a corrupção do 'ó'."""
    with open(caminho_csv, 'r', encoding='utf-8', errors='ignore') as f:
        for i, linha in enumerate(f):
            if termo_busca in linha: return i
    with open(caminho_csv, 'r', encoding='latin-1', errors='ignore') as f:
        for i, linha in enumerate(f):
            if termo_busca in linha: return i
    return 0

def processar_pordata_complexo(caminho_csv, nome_saida):
    linha_skip = encontrar_linha_cabecalho(caminho_csv)
    print(f"📌 A processar: {os.path.basename(caminho_csv)}")
    
    # Lê forçando o separador ponto e vírgula (sep=';')
    try:
        df = pd.read_csv(caminho_csv, skiprows=linha_skip, sep=';', encoding='utf-8')
    except UnicodeDecodeError:
        df = pd.read_csv(caminho_csv, skiprows=linha_skip, sep=';', encoding='latin-1')
        
    # EXTERMINAR AS COLUNAS FANTASMA do Excel (Unnamed)
    df = df.loc[:, ~df.columns.str.contains('Unnamed', case=False, na=False)]
    
    # Corrigir o nome da primeira coluna
    nome_primeira_coluna = df.columns[0]
    df = df.rename(columns={nome_primeira_coluna: "Distrito"})
    
    # Limpar linhas vazias e rodapés institucionais
    df = df.dropna(subset=["Distrito"])
    df = df[~df["Distrito"].str.contains("Fonte|Dados obtidos|Notas|PORDATA", na=False, case=False)]
    
    # EVITAR DUPLA CONTAGEM: Se tiver profissões, ficar só com o "Total"
    if "Tipo de pessoal ao serviço" in df.columns:
        df = df[df["Tipo de pessoal ao serviço"].str.contains("Total", na=False, case=False)]
        df = df.drop(columns=["Tipo de pessoal ao serviço"])
        
    # Limpar e converter todas as colunas numéricas
    colunas_dados = [c for c in df.columns if c != "Distrito"]
    for col in colunas_dados:
        df[col] = df[col].astype(str).str.replace(' ', '').str.replace('-', '0')
        df[col] = df[col].str.replace(',', '.')
        df[col] = pd.to_numeric(df[col], errors='coerce').fillna(0)
        
    # Agrupar por distrito e gravar
    df = df.rename(columns={"Distrito": "Territórios"})
    df_distrito = agregar_por_distrito(df)
    df_distrito = df_distrito.rename(columns={"Territórios": "Distrito"})
    
    caminho_saida = f"data/processed/{nome_saida}"
    df_distrito.to_csv(caminho_saida, index=False)
    print(f"✅ Sucesso! Ficheiro guardado em: {caminho_saida}")
    return df_distrito

# --- EXECUTAR ÓBITOS E HOSPITAIS ---
ficheiro_obitos = "data/raw/obitos.csv"
ficheiro_hospitais = "data/raw/hospitais.csv"

if os.path.exists(ficheiro_obitos):
    df_obitos = processar_pordata_complexo(ficheiro_obitos, "Fact_Obitos.csv")
else:
    print(f"❌ Ficheiro não encontrado: {ficheiro_obitos}")

if os.path.exists(ficheiro_hospitais):
    df_hospitais = processar_pordata_complexo(ficheiro_hospitais, "Fact_Hospitais.csv")
else:
    print(f"❌ Ficheiro não encontrado: {ficheiro_hospitais}")
```

```python
import re
import pandas as pd
import os

def domar_ficheiro_hospitais(caminho_csv, nome_saida):
    print(f"🚑 A iniciar extração de Força Bruta no ficheiro: {os.path.basename(caminho_csv)}")
    
    # 1. Ler o ficheiro bruto
    try:
        df_bruto = pd.read_csv(caminho_csv, sep=';', header=None, encoding='utf-8', dtype=str)
    except UnicodeDecodeError:
        df_bruto = pd.read_csv(caminho_csv, sep=';', header=None, encoding='latin-1', dtype=str)
        
    # 2. CAÇAR OS ANOS
    linha_anos_idx = df_bruto[df_bruto.apply(lambda r: r.astype(str).str.contains('2015').any(), axis=1)].index[0]
    
    linha_anos = df_bruto.iloc[linha_anos_idx].astype(str).tolist()
    novas_colunas = []
    
    for valor in linha_anos:
        match_ano = re.search(r'\d{4}', valor)
        if match_ano: novas_colunas.append(match_ano.group(0))
        else: novas_colunas.append("Descartar")
            
    novas_colunas[0] = "Territórios" 
    df_bruto.columns = novas_colunas
    
    # 3. CAÇAR AS CATEGORIAS (Ficar só com o Total)
    coluna_tipo_idx = None
    for i in range(1, min(5, len(df_bruto.columns))):
        if df_bruto.iloc[:, i].astype(str).str.contains('Total', case=False, na=False).any():
            coluna_tipo_idx = i
            break
            
    if coluna_tipo_idx is not None:
        colunas = list(df_bruto.columns)
        colunas[coluna_tipo_idx] = "Tipo"
        df_bruto.columns = colunas
        df_filtrado = df_bruto[df_bruto["Tipo"].astype(str).str.contains("Total", case=False, na=False)].copy()
    else:
        df_filtrado = df_bruto.copy()
        
    # 4. LIMPAR A CASA
    colunas_finais = ["Territórios"] + [c for c in df_filtrado.columns if c.isdigit()]
    df_limpo = df_filtrado[colunas_finais].copy()
    
    df_limpo = df_limpo.dropna(subset=["Territórios"])
    df_limpo = df_limpo[~df_limpo["Territórios"].str.contains("Fonte|Dados|Notas", case=False, na=False)]
    
    # 5. CONVERTER PARA NÚMEROS (Usando .iloc para evitar o erro do DataFrame!)
    for i in range(len(df_limpo.columns)):
        if df_limpo.columns[i] != "Territórios":
            # Pega na coluna pela posição exata, garante que é uma Series
            serie = df_limpo.iloc[:, i].astype(str)
            serie = serie.str.replace(' ', '').str.replace('-', '0').str.replace(',', '.')
            df_limpo.iloc[:, i] = pd.to_numeric(serie, errors='coerce').fillna(0)
            
    # 5.1 RESOLVER ANOS DUPLICADOS
    # Isolar territórios e números
    df_territorios = df_limpo.iloc[:, 0].copy()
    df_numerico = df_limpo.iloc[:, 1:].copy()
    
    # Agrupar colunas com o mesmo nome (mesmo ano) e somar
    df_numerico = df_numerico.T.groupby(level=0).sum().T
    
    # Juntar tudo novamente
    df_limpo_final = pd.concat([df_territorios, df_numerico], axis=1)
    df_limpo_final.columns = ["Territórios"] + list(df_numerico.columns)
    
    # 6. AGREGAR POR DISTRITO
    df_final = agregar_por_distrito(df_limpo_final)
    df_final = df_final.rename(columns={"Territórios": "Distrito"})
    
    # Gravar o ficheiro
    caminho_saida = f"data/processed/{nome_saida}"
    df_final.to_csv(caminho_saida, index=False)
    
    print(f"✅ SUCESSO ABSOLUTO! Ficheiro guardado em: {caminho_saida}")
    print(f"Colunas finais: {df_final.columns.tolist()[:6]}...\n")
    return df_final

# 7. UNPIVOT (Passar de formato Largo para Longo)
    colunas_anos = [c for c in df_final.columns if c != "Distrito"]
    df_final = df_final.melt(
        id_vars=["Distrito"], 
        value_vars=colunas_anos, 
        var_name="Ano", 
        value_name="Total_Profissionais"
    )
    
    # Agora sim, gravar o ficheiro...
# --- EXECUTAR SÓ PARA OS HOSPITAIS ---
ficheiro_hospitais = "data/raw/hospitais.csv"
if os.path.exists(ficheiro_hospitais):
    df_hospitais = domar_ficheiro_hospitais(ficheiro_hospitais, "Fact_Hospitais.csv")
else:
    print(f"❌ Ficheiro não encontrado: {ficheiro_hospitais}")
```
