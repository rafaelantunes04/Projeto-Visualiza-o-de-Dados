```python
import os
import re
import pandas as pd
import itertools

def domar_obitos_lote(caminho_csv):
    print(f"A processar: {os.path.basename(caminho_csv)}")
    try:
        df_bruto = pd.read_csv(caminho_csv, sep=';', header=None, encoding='utf-8', dtype=str)
    except UnicodeDecodeError:
        df_bruto = pd.read_csv(caminho_csv, sep=';', header=None, encoding='latin-1', dtype=str)
        
    idx_doencas = None
    for idx, row in df_bruto.iterrows():
        if len(row) > 2 and str(row[2]).strip() == "Diabetes":
            idx_doencas = idx
            break
            
    if idx_doencas is None:
        print("ERRO: Não encontrou linha de Doenças.")
        return pd.DataFrame()
        
    idx_anos = idx_doencas + 2
    linha_doencas = df_bruto.iloc[idx_doencas].fillna('').astype(str).tolist()
    linha_anos = df_bruto.iloc[idx_anos].fillna('').astype(str).tolist()

    mapa_colunas = {}
    doenca_atual = "Desconhecida"
    for i in range(2, len(linha_doencas)):
        val_doenca = linha_doencas[i].strip()
        val_ano = linha_anos[i].strip()
        if val_doenca not in ['nan', '', 'None'] and len(val_doenca) > 3:
            doenca_atual = val_doenca
        if re.match(r'\d{4}', val_ano):
            mapa_colunas[i] = {"Causa": doenca_atual, "Ano": val_ano}

    chaves_municipios = {k.lower(): k for k in DICIONARIO_DISTRITOS.keys()}
    registos = []

    for row_idx in range(idx_anos + 1, len(df_bruto)):
        row = df_bruto.iloc[row_idx].fillna('').astype(str)
        municipio_encontrado = None
        for col_m in range(3):
            valor_limpo = row[col_m].strip().lower()
            if valor_limpo in chaves_municipios:
                municipio_encontrado = chaves_municipios[valor_limpo] 
                break

        if municipio_encontrado:
            for col_idx, info in mapa_colunas.items():
                val_num = row[col_idx].replace(' ', '').replace(',', '.').replace('x', '0').replace('X', '0').replace('-', '0')
                if val_num in ['', ',,,', 'nan']:
                    val_num = '0'
                try:
                    registos.append({
                        "Territórios": municipio_encontrado,
                        "Ano": info["Ano"],
                        "Causa_Morte": info["Causa"],
                        "Obitos": float(val_num)
                    })
                except ValueError:
                    pass

    df_longo = pd.DataFrame(registos)
    if df_longo.empty: return df_longo

    df_longo["Distrito"] = df_longo["Territórios"].map(DICIONARIO_DISTRITOS)
    df_longo = df_longo.dropna(subset=["Distrito"])
    
    df_final = df_longo.groupby(["Distrito", "Ano", "Causa_Morte"])["Obitos"].sum().reset_index()
    return df_final

# --- EXECUÇÃO PRINCIPAL ---
ficheiro1 = "data/raw/Óbitos de residentes em Portugal por algumas causas de morte 1.csv"
ficheiro2 = "data/raw/Óbitos de residentes em Portugal por algumas causas de morte 2.csv"

# 1. Processar os dois datasets 
df1 = domar_obitos_lote(ficheiro1) if os.path.exists(ficheiro1) else pd.DataFrame()
df2 = domar_obitos_lote(ficheiro2) if os.path.exists(ficheiro2) else pd.DataFrame()

# 2. Unificar os dados (Merge)
print("\n A unificar Dados...")
df_obitos = pd.concat([df1, df2], ignore_index=True)
df_obitos = df_obitos.drop_duplicates(subset=["Distrito", "Ano", "Causa_Morte"]).sort_values(by=["Distrito", "Causa_Morte", "Ano"])

# 3. INTERPOLAÇÃO PARA PREENCHER OS ANOS VAZIOS (Ex: 2002 a 2008)
print("A aplicar Interpolação Linear para preencher anos em falta...")

df_obitos["Ano"] = df_obitos["Ano"].astype(int)
ano_min, ano_max = df_obitos["Ano"].min(), df_obitos["Ano"].max()
distritos_unicos = df_obitos["Distrito"].unique()
causas_unicas = df_obitos["Causa_Morte"].unique()

# Criar um esqueleto perfeito (Distrito x Causa x Ano)
esqueleto = list(itertools.product(distritos_unicos, causas_unicas, range(ano_min, ano_max + 1)))
df_esqueleto = pd.DataFrame(esqueleto, columns=["Distrito", "Causa_Morte", "Ano"])

# Juntar os dados reais (os anos em falta ficarão como NaN)
df_completo = pd.merge(df_esqueleto, df_obitos, on=["Distrito", "Causa_Morte", "Ano"], how="left")

# Agrupar por Distrito e Doença para calcular a reta de crescimento
df_completo["Obitos"] = (
    df_completo.groupby(["Distrito", "Causa_Morte"])["Obitos"]
    .apply(lambda x: x.interpolate(method='linear'))
    .reset_index(level=[0,1], drop=True)
)

# Limpar, arredondar a inteiro e ordenar para o formato final
df_completo["Obitos"] = df_completo["Obitos"].fillna(0).round(0).astype(int)
df_completo = df_completo.sort_values(by=["Distrito", "Ano", "Causa_Morte"])

# Gravar o resultado final
caminho_final = "data/processed/Obitos_por_alguma_Causas.csv"
os.makedirs("data/processed", exist_ok=True)
df_completo.to_csv(caminho_final, index=False)

print(f"SUCESSO! Dataset final, longo e interpolado gravado em: {caminho_final}")
print("\nAmostra do preenchimento mágico:")
print(df_completo.head())
```

```python
import os
import re
import pandas as pd
import itertools

def domar_hospitais_mestre(caminho_csv, nome_saida):
    print(f"A processar o dataset: {os.path.basename(caminho_csv)}")
    
    try:
        df_bruto = pd.read_csv(caminho_csv, sep=';', header=None, encoding='utf-8', dtype=str)
    except UnicodeDecodeError:
        df_bruto = pd.read_csv(caminho_csv, sep=';', header=None, encoding='latin-1', dtype=str)
        
    # 2. CAÇAR AS LINHAS CHAVE (Categorias e Anos)
    idx_categorias = None
    idx_anos = None
    
    for i, row in df_bruto.iterrows():
        linha_str = ' '.join(row.fillna('').astype(str)).lower()
        if 'total' in linha_str and 'enfermeiros' in linha_str:
            idx_categorias = i
        if '1999' in linha_str and '2023' in linha_str and 'anos' in linha_str:
            idx_anos = i
            
        if idx_categorias is not None and idx_anos is not None:
            break
            
    if idx_categorias is None or idx_anos is None:
        print("ERRO: Não foi possível identificar as linhas de cabeçalho.")
        return pd.DataFrame()
        
    linha_cats = df_bruto.iloc[idx_categorias].fillna('').astype(str).tolist()
    linha_anos = df_bruto.iloc[idx_anos].fillna('').astype(str).tolist()
    
    # 3. ISOLAR AS COLUNAS DO "TOTAL"
    colunas_total = {}
    capturar = False
    
    for i in range(2, len(linha_cats)):
        cat_val = linha_cats[i].strip().lower()
        
        if 'total' in cat_val:
            capturar = True
        elif cat_val != '' and capturar:
            # Chegámos à próxima categoria (ex: Médicos), paramos de capturar!
            capturar = False
            break 
            
        if capturar:
            ano_str = linha_anos[i].strip()
            match = re.search(r'\b(19|20)\d{2}\b', ano_str)
            if match:
                colunas_total[i] = match.group(0) # Guarda a posição da coluna e o ano correspondente

    # 4. EXTRAÇÃO E LIMPEZA
    chaves_municipios = {k.lower(): k for k in DICIONARIO_DISTRITOS.keys()}
    registos = []

    for row_idx in range(idx_anos + 1, len(df_bruto)):
        row = df_bruto.iloc[row_idx].fillna('').astype(str)
        
        municipio_encontrado = None
        for col_m in range(3):
            valor_limpo = row[col_m].strip().lower()
            if valor_limpo in chaves_municipios:
                municipio_encontrado = chaves_municipios[valor_limpo]
                break
                
        if municipio_encontrado:
            for col_idx, ano in colunas_total.items():
                val_bruto = row[col_idx]
                
                # A MAGIA DA LIMPEZA: Remove 'Pro', espaços, e converte '//' num vazio
                val_num = re.sub(r'[^\d]', '', str(val_bruto))
                
                if val_num == '':
                    val_num = '0' # Municípios sem hospital ou sem dados reportados
                    
                registos.append({
                    "Territórios": municipio_encontrado,
                    "Ano": ano,
                    "Total_Profissionais": float(val_num)
                })

    df_longo = pd.DataFrame(registos)
    if df_longo.empty: return df_longo

    # 5. MAPEAMENTO DE DISTRITOS
    df_longo["Distrito"] = df_longo["Territórios"].map(DICIONARIO_DISTRITOS)
    df_longo = df_longo.dropna(subset=["Distrito"])
    
    # Agrupar a soma dos profissionais por Distrito e Ano
    df_agrupado = df_longo.groupby(["Distrito", "Ano"])["Total_Profissionais"].sum().reset_index()
    
    return df_agrupado


# --- EXECUÇÃO, INTERPOLAÇÃO E ALINHAMENTO TEMPORAL ---
caminho_hosp = "data/raw/hospitais.csv"
if os.path.exists(caminho_hosp):
    df_hosp = domar_hospitais_mestre(caminho_hosp, "Fact_Hospitais.csv")
    
    if not df_hosp.empty:
        print("A aplicar Interpolação Linear para preencher os anos ausentes...")
        df_hosp["Ano"] = df_hosp["Ano"].astype(int)
        
        ano_min = df_hosp["Ano"].min()
        ano_max = df_hosp["Ano"].max()
        distritos = df_hosp["Distrito"].unique()
        
        # O esqueleto que vai forçar a criação dos anos invisíveis
        esqueleto = list(itertools.product(distritos, range(ano_min, ano_max + 1)))
        df_esq = pd.DataFrame(esqueleto, columns=["Distrito", "Ano"])
        
        # Juntar dados e interpolar matematicamente os vazios
        df_completo = pd.merge(df_esq, df_hosp, on=["Distrito", "Ano"], how="left")
        
        df_completo["Total_Profissionais"] = (
            df_completo.groupby("Distrito")["Total_Profissionais"]
            .apply(lambda x: x.interpolate(method='linear'))
            .reset_index(level=0, drop=True)
        )
        
        df_completo["Total_Profissionais"] = df_completo["Total_Profissionais"].fillna(0).round(0).astype(int)
        
        # 6. ALINHAMENTO COM OS ÓBITOS (Eliminar 1999 e 2000)
        df_completo = df_completo[df_completo["Ano"] >= 2001].sort_values(by=["Distrito", "Ano"]).reset_index(drop=True)
        
        os.makedirs("data/processed", exist_ok=True)
        caminho_saida = "data/processed/Pessoal_Hospitais.csv"
        df_completo.to_csv(caminho_saida, index=False)
        
        print(f"SUCESSO! Ficheiro gerado com precisão: {caminho_saida}")
        print("\nAmostra dos resultados (verifica os anos 2001 a 2005 da Guarda):")
        print(df_completo.head(5).to_string(index=False))
else:
    print(f"Ficheiro não encontrado: {caminho_hosp}")
```
